# Conditionals: `if`...`elif`...`else`

`if` is a method that allows responses (statements) to occur if certain conditions are met.

__Syntax:__ 
\[__Note:__ There are four(4) spaces on the line after the colon (:). This lets the Python interpreter know the items with that spacing all belongs to the "_scope_" of that conditions response.\]

```
if (condition is true):
    statement
    statement_b
elif (optional condition is true):
    statement_2
else:
    statement_3
```
---
## Example 1:

### A single `if` statement example

```
x = 3
if (x > 1): print(x,"is greater than 1") # The 1 line response statement can work like this
if (x > 1): 
    print(x,"is greater than 1") # Or like this on seperate lines
```

---
## Example 2a:

### A multiple if statement example (Is a car older than a given year)
__When there are multiple `if` statements, each is checked unless `elif` is used__

```
car_year = 1979
if (car_year < 1990): 
    print("The car is older than 1990")
if (car_year < 1980): 
    print("The car is older than 1980")
if (car_year == 1979): 
    print("The car was made in 1979")
```
## Example 2b:

### Now let's try the same example as a elif statement example

```
car_year = 1979
if (car_year < 1990): 
    print("The car is older than 1990")
elif (car_year < 1980): 
    print("The car is older than 1980")
elif (car_year == 1979): 
    print("The car was made in 1979")
```
__Notice only one response was printed with the `if...elif`. This is becasue that method exits after completing 1 true conditions response.__

## Example 3:

## A `if...elif...else` statement example (Does the car fall in a certian decade?)
\[__Note:__ The `and` is used to say the first and the second condition must be true. You could also use `or` instead.\] 

```
car_year = 1979
if (car_year < 2000 and car_year >= 1990): 
    print("The car was made in the 90's")
elif (car_year < 1990 and car_year >= 1980): 
    print("The car was made in the 80's")
elif (car_year < 1980 and car_year >= 1970): 
    print("The car was made in the 70's")
else:
    print("Congratulations if that thing is still running!")
```
