# Importing a Module

Let's use the example that we want to use the Pythagorean Theorem $a^2+b^2=c^2$ to solve for "$c$" where `a=5` and `b=6`. 
so $c=\sqrt(a^2 +b^2)$ or $c=\sqrt{5^2 +6^2}$
The built in math functionality of Python does not include $\sqrt[y]{x}$ nor $x^y$. For these functions we will add/import the `math` module as follows.

## Example

```python
import math # Adds the functionality of the math module
```

\[__Note:__ If the import of the module return an error not due to spelling, it would need to be installed by your "Module Manager". With Anaconda you would use a command such as "conda install math" or "pip3 install math". Further resources about such issues can be found online.\]

Now lets look and see what functions are available in the `math` module using `dir(math)`

If a function with the name `__doc__` is avaialable (_and the developer added information into it_), it can be called to get information about the function. Below is an example for both `sqrt` and `pow`.  

Now that we have (1) imported the `math` module and (2) found the needed functionality, we can solve our inital problem. 

```python
a = 5
b = 6
c = math.sqrt(pow(a,2)+pow(b,2))
print("c is equal to",c)
```
