# Using Markdown in Jupyter Notebooks

Markdown is a minimal language used to format text. It is most often used on sites such as Github, Wikipedia, and of course Jupyter Notebooks. 

## To render/run any markdown/code in a cell in Jupyter you must either 
* <span style="color:blue">**Press [Shift]+[Enter]**</span> 
* <span style="color:blue">**Click the “▶| Run” button in the Toolbar**</span>
